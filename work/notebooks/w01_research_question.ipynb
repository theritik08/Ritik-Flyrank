## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Unit of analysis is a single page.

The decision: which pages a reviewer should actually spend time on first,
out of way more pages than anyone has time to check manually. Right now
that's probably just gut feeling or whoever complains loudest.

Who acts on it: whoever's doing content/SEO review — they'd get a shortlist
instead of scrolling a spreadsheet.

The action: reviewer looks at the page, maybe rewrites the title or meta
description, or checks if it's even still answering what people are
searching for.

Cost if I get it wrong: if I flag a page that's actually fine, that's
wasted review time — and worse, it means the real problem page didn't get
looked at instead. If I miss an actual underperformer, nobody notices and
the client just keeps losing clicks they could've had.

Why this needs data instead of just eyeballing it: because raw CTR is
mostly explained by position, a person skimming a list of low-CTR pages
would mostly just be finding low-position pages, not actual problems.
Adjusting for position first is the part a person wouldn't naturally do
by hand across thousands of rows.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd, numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/theritik08/Ritik-Flyrank/refs/heads/main/data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")

# Number 1: CTR by position tier
visible = df[df["impressions_90d"] >= 100]
tier_ctr = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print("\nMean CTR by position tier:")
print(tier_ctr.round(4).to_string())

# Number 2: pages noticeably below their own tier's average CTR
visible500 = df[df["impressions_90d"] >= 500].copy()
tier_mean = visible500.groupby("position_tier")["ctr"].transform("mean")
visible500["ctr_gap_pct"] = (visible500["ctr"] - tier_mean) / tier_mean

underperformers = visible500[visible500["ctr_gap_pct"] <= -0.5]
print(f"\nVisible pages (impressions_90d >= 500): {len(visible500):,}")
print(f"Pages at least 50% below their tier's average CTR: {len(underperformers):,} "
      f"({len(underperformers)/len(visible500):.1%})")

print(f"Total impressions in underperforming pages: {underperformers['impressions_90d'].sum():,.0f}")
```
